```
def f1(sal):
  if sal == "<=50K":
        return 0
  else:
        return 1
df["income_modified"]=df["income"].apply(f1)
df.head(1)
```

```
import pandas as pd

data = pd.DataFrame({
    "Salary": [25000, 27000, 28000, 30000, 32000, 35000,
               38000, 40000, 42000, 45000, 48000, 50000]
})
```

```
import pandas as pd
data["Salary"].hist()
```

using matplotlib

```
import matplotlib.pyplot as plt
plt.hist(data["Salary"])
```

using seaborn

```
import seaborn as sns
sns.histplot(data["Salary"])
```

use of bins:- use used to avoid number of bars for ech value in column by grouping them thus makes graph earier to read and avoid messy

```
plt.hist(data["Salary"], bins=2)
plt.show()

plt.hist(data["Salary"], bins=4)
plt.show()

plt.hist(data["Salary"], bins=8)
plt.show()

plt.hist(data["Salary"], bins=12)
plt.show()
```

ans:-The easiest graph to understand is usually the one with a moderate number of bins (for this small dataset, around 4–8 bins). It balances detail and readability.
